```python
#crop

import numpy as np
import cv2
image=cv2.imread(r"C:\Users\arunf\OneDrive\Documents\Data Science\AI\messi.jpg")
width,height=250,350
point1=np.float32([[111,400],[200,200],[230,482],[230,440]])
point2=np.float32([[0,0],[width,0],[0,height],[width,height]])
matrix=cv2.getPerspectiveTransform(point1,point2)
output=cv2.warpPerspective(image,matrix,(width,height))
cv2.imshow("image",image)
cv2.imshow("output",output)
k=cv2.waitKey()
if k==27:
    cv2.destroyAllWindows()
```

```python
import cv2
img=cv2.imread(r"C:\Users\arunf\Downloads\lock screen wallpaper.jpg")
print('original dimensions : ',img.shape)
scale=200
width=int(img.shape[1]*scale/100)
height=int(img.shape[0]*scale/100)
dim=(width,height)
resized=cv2.resize(img,dim,interpolation=cv2.INTER_AREA)
print('resized dimension',resized.shape)
cv2.imshow('resized image',resized)
k=cv2.waitKey()
if k==27:
    cv2.destroyAllWindows()
```
