```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
ds=pd.read_csv(r"C:\Users\arunf\OneDrive\Documents\Data Science\Machine Learning\DataSet\diabetes (3) (2).csv")
```

```python
ds
```

```python
ds.isnull().sum()
```

```python
train,test=train_test_split(ds,test_size=0.2)
```

```python
trainx=train.iloc[:,:8]
trainy=train.iloc[:,8]
testx=test.iloc[:,:8]
testy=test.iloc[:,8]
```

```python
trainx
```

```python
trainy
```

```python
rfc=RFC(n_jobs=8,n_estimators=20,oob_score=True,criterion='entropy')
```

```python
pred=rfc.fit(trainx,trainy).predict(testx)
cm=confusion_matrix(pred,testy)
acc=np.mean(pred==testy.values.flatten())
```
