```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
ds=pd.read_csv(r"C:\Users\arunf\OneDrive\Documents\Data Science\Machine Learning\DataSet\kc_house_data.csv")
```

```python
ds
```

```python
ds=ds.drop('id',axis=1)
ds=ds.drop('date',axis=1)
```

```python
ds
```

```python
train,test=train_test_split(ds,test_size=0.2)
```

```python
trainx=train.iloc[:,1:]
trainy=train.iloc[:,0]
testx=test.iloc[:,1:]
testy=test.iloc[:,0]
```

```python
trainx
```

```python
trainy
```

```python
lr=LinearRegression()
```

```python
pred=lr.fit(trainx,trainy).predict(testx)
```

```python
score=r2_score(pred,testy)
```
