```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn import preprocessing
```

```python
ds=pd.read_csv(r"C:\Users\arunf\OneDrive\Documents\Data Science\Machine Learning\DataSet\combines (1).csv")
```

```python
ds
```

```python
le=preprocessing.LabelEncoder()
```

```python
ds['model']=le.fit_transform(ds['model'].astype(str))
ds['transmission']=le.fit_transform(ds['transmission'].astype(str))
ds['fuelType']=le.fit_transform(ds['fuelType'].astype(str))
ds['brand']=le.fit_transform(ds['brand'].astype(str))
```

```python
ds
```

```python
ds.isnull().sum()
```

```python
train,test=train_test_split(ds,test_size=0.2)
```

```python
trainx=train.iloc[:,[0,1,3,4,5,6,7,8,9]]
trainy=train.iloc[:,2]
testx=test.iloc[:,[0,1,3,4,5,6,7,8,9]]
testy=test.iloc[:,2]
```

```python
lr=LinearRegression()
```

```python
pred=lr.fit(trainx,trainy).predict(testx)
```
